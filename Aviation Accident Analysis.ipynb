```python
# We are going to drop columns that have roughly more than 25% of their data missing
drop_columns=['Latitude', 'Longitude', 'Airport.Code', 'Airport.Name', 'Aircraft.Category', 'FAR.Description',
                   'Schedule', 'Air.carrier', 'Broad.phase.of.flight']
data_clean=Data.drop(columns=drop_columns)
```

```python
#Change some column titles to more managable ones 
new_column_name={'Event.Id': 'ID', 'Investigation.Type': 'Type', 'Event.Date':'Date', 'Injury.Severity':'Injury_Severity', 
                    'Aircraft.damage':'Damage_type', 'Number.of.Engines':'Engines', 'Purpose.of.flight':'Flight_Purpose',
                    'Total.Fatal.Injuries':'Fatal_Injuries', 'Total.Serious.Injuries':'Serious_Injuries',
                   'Total.Minor.Injuries':'Minor_Injuries', 'Total.Uninjured':'Uninjured', 'Weather.Condition':'Weather',}
data_clean.rename(columns=new_column_name, inplace=True)
```

```python
data_clean.head()
```

While exploring the data above it came to my attention that columns 'Injury.Severity' and 'Total.Fatal.Injuries' provide similar information but 'Total.Fatal.Injuries' has many missing data. By extracting data from these two columns, I create a new Fatality column with less missing numerical data than 'Total.Fatal.Injuries'

```python
# Extract numbers from 'Injury_Severity' into 'Fatality', or keep original value if not available
data_clean['Fatality'] = data_clean['Injury_Severity'].str.extract(r'\((\d+)\)', expand=False)

# Replace specific textual values with 0
data_clean['Fatality'].replace({'Non-Fatal': 0, 'Minor': 0, 'Serious': 0, 'Incident': 0}, inplace=True)

# Update 'Fatality' where 'Injury_Severity' is 'Fatal'
data_clean['Fatality'] = data_clean['Fatality'].fillna(data_clean['Fatal_Injuries'])
data_clean['Fatality'] = data_clean['Fatality'].where(data_clean['Fatality'] != 'Fatal', data_clean['Fatal_Injuries'])

# Replace 'Unavailable' with NaN
data_clean['Fatality'].replace('Unavailable', np.nan, inplace=True)

# Convert 'Fatality' to numeric values, forcing errors to NaN
data_clean['Fatality'] = pd.to_numeric(data_clean['Fatality'], errors='coerce')

# Set display format for floats
pd.options.display.float_format = '{:.0f}'.format
```

```python
# We can drop Fatal_injuries column now
data_clean.drop(columns=['Fatal_Injuries'],inplace=True)
```

```python
# changing date type to the appropriate format and creating a column for seasons
data_clean['Date']=pd.to_datetime(data_clean['Date'],format='%Y-%m-%d')
data_clean['Month']=data_clean['Date'].dt.month
```

```python
#Adding season name according to the month number
Season={ 12: 'Winter', 1: 'Winter', 2: 'Winter',
    3: 'Spring', 4: 'Spring', 5: 'Spring',
    6: 'Summer', 7: 'Summer', 8: 'Summer',
    9: 'Fall', 10: 'Fall', 11: 'Fall'
}
data_clean['Season']=data_clean['Month'].map(Season)
```

```python
#Adding year column 
data_clean['Year']=data_clean['Date'].dt.year
```

I create new columns using Location for future geographical analysis and visualisations. 
This was easier on Jupyter notebook by pip install us & import us, but it did not work on Kaggle
So I ended up creating a list of valid US states codes

```python
valid_state_codes = [
    'AL', 'AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA', 
    'HI', 'ID', 'IL', 'IN', 'IA', 'KS', 'KY', 'LA', 'ME', 'MD', 
    'MA', 'MI', 'MN', 'MS', 'MO', 'MT', 'NE', 'NV', 'NH', 'NJ', 
    'NM', 'NY', 'NC', 'ND', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 
    'SD', 'TN', 'TX', 'UT', 'VT', 'VA', 'WA', 'WV', 'WI', 'WY'
]
def extract_city_state(location):
    if pd.notna(location):
        location = location.strip()
        last_two_chars = location[-2:].upper()
        if last_two_chars in valid_state_codes:
            return location[:-3].strip(), last_two_chars
        else:
            return location, "Not Applicable"
    else:
        return np.nan, np.nan
    
data_clean[['City', 'State']] = data_clean['Location'].apply(extract_city_state).apply(pd.Series)
```

```python
# getting rid of trailing commas in City column
data_clean['City']=data_clean['City'].str.rstrip(',')
```

## Conclusion and Recommendations

The decline in aviation accidents since 1982 is promising, yet states like California, Texas, and Florida still face higher accident rates. To address this, it's crucial to continuously improve safety protocols and regulations. Increased surveillance and enforcement in high-accident states can help reduce risks.

Accidents often occur in good weather when pilots might become complacent. To combat this, we should emphasize safe flying practices and discourage risky behaviors through training and awareness.

Personal flights contribute significantly to accidents. Promoting thorough education and a safety-first culture among personal pilots is essential. Ensuring pilots are well-prepared and aware of risks, regardless of weather, will help decrease accidents.
